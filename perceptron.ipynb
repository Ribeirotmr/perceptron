```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
df = pd.read_csv("perceptron.csv")
X = df[['Feature1', 'Feature2', 'Feature3']].values.tolist()
Y = df['Class_Label'].values
Y = np.where(Y == 1, 1, -1)


def percepton(X, Y):

    linha = len(X)
    coluna = len(X[0])
    tamanho = coluna
    b = 0

    u = [0] * tamanho

    t_interacoes = linha * coluna + linha

    for _ in range(t_interacoes):
        qtd_vezes = 0

        for i in range(linha):
            soma = 0
            for j in range(tamanho):
                soma += u[j] * X[i][j]
            y = soma + b

            if y >= 0:
                y_sinal = 1
            else:
                y_sinal = -1

            if Y[i] != y_sinal:
                for k in range(tamanho):
                    u[k] += Y[i] * X[i][k]
                b += Y[i]
                qtd_vezes += 1

        if qtd_vezes == 0:
            print("Convergiu")
            break
    return u, b


u, b = percepton(X, Y)

print("Vetor de pesos:", u)
print("Bias:", b)


# Calcular acurácia
acertos = 0
for i in range(len(X)):
    soma = 0
    for j in range(len(u)):
        soma += u[j] * X[i][j]
    y_pred = soma + b
    classe_pred = 1 if y_pred >= 0 else -1
    if classe_pred == Y[i]:
        acertos += 1

acuracia = acertos / len(X) * 100
print(f"\nAcurácia: {acuracia:.2f}%")
```

```python
X_np = np.array(X)
f2 = X_np[:, 1] 
f3 = X_np[:, 2] 

plt.figure(figsize=(10, 5))
plt.scatter(f2[Y == 1], f3[Y == 1], color='blue', label='Classe 1', edgecolors='k', s=70)
plt.scatter(f2[Y == -1], f3[Y == -1], color='red', label='Classe 0', edgecolors='k', s=70)

x_vals = np.linspace(min(f2), max(f2), 100)
y_vals = (-u[1] * x_vals - u[0] * 1.0 - b) / u[2]

plt.plot(x_vals, y_vals, '--g', label='Fronteira de Decisão', linewidth=2)

plt.title(f"Perceptron 2D (Acurácia: {acuracia:.2f}%)")
plt.xlabel("Feature 2")
plt.ylabel("Feature 3")
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.ylim(min(f3) - 0.1, max(f3) + 0.1) 

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 1. Configuração da figura 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 2. Plotar os pontos dos dados
X_np = np.array(X)
colors = ['blue' if y == 1 else 'red' for y in Y]
ax.scatter(X_np[:, 0], X_np[:, 1], X_np[:, 2], c=colors, s=50, edgecolors='k', label='Dados')

# 3. Criar a superfície do hiperplano
# Equação: u0*x + u1*y + u2*z + b = 0  =>  z = (-u0*x - u1*y - b) / u2
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 10),
                     np.linspace(ylim[0], ylim[1], 10))

# Calculando o Z correspondente no plano
zz = (-u[0] * xx - u[1] * yy - b) / u[2]

# Plotar o plano
ax.plot_surface(xx, yy, zz, alpha=0.3, color='green', label='Decisão')

# 4. Estilização
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('Visualização do Perceptron 3D')

plt.show()
```
